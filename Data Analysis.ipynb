```python
def summary_report_numericals(data):   
    summary = pd.DataFrame(data.dtypes,columns = ['data_type'])
    summary = summary.reset_index()
    summary['column_name'] = summary['index']
    ## Identifying the missing values
    summary['missing'] = data.isnull().sum().values
    summary['missing%'] = (data.isnull().sum()/len(data)*100).values
    summary['uniques'] = data.nunique().values
    summary['unique%'] = (data.nunique()/len(data)*100).values
    return summary
```

```python
### CLT by defining function
def central_limit_theorem(data,col,N,n):
    sample_mean=[]
    for i in range(N):
        sample_mean.append(np.mean(random.sample(list(df['age']),n)))
    return sample_mean
def central_limit_theorem_graph(data,col,N,n):
    sample_mean=central_limit_theorem(data,col,N,n)
    mean_diff=np.mean(data[col])-np.mean(sample_mean)
    plt.text(0.7,0.9,f'mean:{np.mean(sample_mean)}',transform= plt.gca().transAxes)
    plt.text(0.7,0.8,f'difference:{mean_diff}',transform=  plt.gca().transAxes)
    plt.hist(sample_mean)
    plt.show()
```

```python
### using function
def Z_confidence_interval(data,col,n,conf):
    sample_mean=np.mean(random.sample(list(df['age']),n))
    standard_error=np.std(df['age'])/np.sqrt(n) ### sigma/sqrt(n)
    z_star=stats.norm.ppf(((1-conf_interval)/2)+conf_interval)
    CI=sample_mean-(z_star*standard_error),sample_mean+(z_star*standard_error)
    return CI[0],CI[1]
```

```python
Z_confidence_interval(df,'age',1000,0.99)
```

```python
def t_confidence_interval_sample_mean(data,column,n,conf):
    sample=random.sample(list(data[column]),n)
    sample_mean=np.mean(sample)
    sample_std=np.std(sample)
    
    standard_error=sample_std/np.sqrt(n-1)
    t_score=stats.t.ppf(((1-conf)/2)+conf,n-1)
    CI=sample_mean-(t_score*standard_error),sample_mean+(t_score*standard_error)
    return CI[0],CI[1]
```

```python
t_confidence_interval_sample_mean(df,'age',29,0.95)
```

```python
from scipy.stats import chi2_contingency

def calculate_Chi2(data,columns_0,columns_1):
    data_crosstab = pd.crosstab(data[columns_0], data[columns_1], margins = False) 
    stat, p, dof, expected = chi2_contingency(data_crosstab)
    alpha = 0.05
    print('significance=%.3f, p=%.3f' % (alpha, p))
    if p <= alpha:
        print(f' Given variables [{columns_0},{columns_1}] are Dependent')
    else:
        print(f'Given variables [{columns_0},{columns_1}] are Independent')
    return stat, p
```

```python
from itertools import permutations
for columns in list(permutations(data.columns, 2)):    
    calculate_Chi2(data,columns[0],columns[1])
```
